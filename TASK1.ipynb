<b>TASK 1:
create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.</b>


```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulated population distribution data (age group: population in thousands)
data = {
    "Age Group": [
        "0-4", "5-9", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39",
        "40-44", "45-49", "50-54", "55-59", "60-64", "65-69", "70-74", "75-79", "80+"
    ],
    "Population (Thousands)": [
        12000, 11500, 11000, 10500, 10000, 9500, 9000, 8500,
        8000, 7500, 7000, 6500, 6000, 5500, 5000, 4500, 4000
    ]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plotting the histogram
plt.figure(figsize=(12, 8))
plt.bar(df["Age Group"], df["Population (Thousands)"], color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Population (Thousands)')
plt.title("Population Distribution by Age in India (2023)")
plt.xticks(rotation=45)
#plt.grid(axis='y', linestyle='--', alpha=0.7)

# Displaying the plot
plt.tight_layout()
plt.show()
```
